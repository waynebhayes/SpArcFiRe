```julia
include("MyFunctions.jl");
using Gadfly, Colors
ACCTe = load("numbers.jld")["ACCTe"]
ACCTr = load("numbers.jld")["ACCTr"]
RMSETe = load("numbers.jld")["RMSETe"]
RMSETr = load("numbers.jld")["RMSETr"]
PCTe = load("numbers.jld")["PCTe"]
PCTr = load("numbers.jld")["PCTr"]
palette = rand(distinguishable_colors(1000),20)
names = [string(trees) for trees in 1:20]
set_default_plot_size(30cm, 30cm);
```

```julia
n=5
paccTr = plot(Guide.title("Random Forests chirality prediction (on Training Set)"), Guide.ylabel("Accuracy %"), Guide.xlabel(nothing),
#Guide.manual_color_key("Number of trees", names[1:n], palette[1:n]),
[layer(x=collect(1:n), y=(ACCTr[[2,3,6,9,10],features]+rand(collect(linspace(.1,.8,10000)),n)), 
    Geom.point, Geom.line, Theme(default_color=palette[features],major_label_font_size=14pt, minor_label_font_size=20pt, key_label_font_size=.4cm))
    for features in 1:n]...)
paccTe = plot(Guide.ylabel(nothing), Guide.xlabel(nothing), Guide.title("Random Forests chirality prediction (on Test Set)"), #Guide.xlabel("Number of Features"),
    Guide.manual_color_key("Trees", names[1:n], palette[1:n]),
[layer(x=collect(1:n), y=(ACCTe[1:n,features]+rand(collect(linspace(.1,.8,10000)),n)), 
    Geom.point, Geom.line, Theme(default_color=palette[features],major_label_font_size=14pt, minor_label_font_size=20pt, key_label_font_size=.4cm))
    for features in 1:n]...)

ppcTr=plot(Guide.ylabel("Pearson Correlation"), Guide.xlabel(nothing), #Guide.title("Rnadom Forests chirality prediction (on Training Set)"), Guide.xlabel("Number of Features"),
    #Guide.manual_color_key("Number of trees", names[1:n], palette[1:n]),
    [layer(x=collect(1:n), y=(PCTr[1:n,features]+rand(collect(linspace(.1,.8,10000)),n)), 
    Geom.point, Geom.line, Theme(default_color=palette[features],major_label_font_size=14pt, minor_label_font_size=20pt, key_label_font_size=.4cm))
    for features in 1:n]...)
ppcTe=plot(Guide.ylabel(nothing), Guide.xlabel(nothing), #Guide.title("Rnadom Forests chirality prediction (on Training Set)"), 
    Guide.manual_color_key("Trees", names[1:n], palette[1:n]),
    [layer(x=collect(1:n), y=(PCTe[1:n,features]+rand(collect(linspace(.1,.8,10000)),n)), 
    Geom.point, Geom.line, Theme(default_color=palette[features],major_label_font_size=14pt, minor_label_font_size=20pt, key_label_font_size=.4cm))
    for features in 1:n]...)

prmseTr = plot(Guide.ylabel("RMSE"), Guide.xlabel("Number of Features"), #Guide.title("Rnadom Forests chirality prediction (on Training Set)"), 
    #Guide.manual_color_key("Number of trees", names[1:n], palette[1:n]),
    [layer(x=collect(1:n), y=(RMSETr[1:n,features]+rand(collect(linspace(.1,.8,10000)),n)), 
    Geom.point, Geom.line, Theme(default_color=palette[features],major_label_font_size=14pt, minor_label_font_size=20pt, key_label_font_size=.4cm))
    for features in 1:n]...)
prmseTe = plot(Guide.ylabel(nothing),  Guide.xlabel("Number of Features"), #Guide.title("Rnadom Forests chirality prediction (on Training Set)"),
    Guide.manual_color_key("Trees", names[1:n], palette[1:n]),
    [layer(x=collect(1:n), y=(RMSETr[1:n,features]+rand(collect(linspace(.1,.8,10000)),n)), 
    Geom.point, Geom.line, Theme(default_color=palette[features],major_label_font_size=14pt, minor_label_font_size=20pt, key_label_font_size=.4cm))
    for features in 1:n]...)

finalPlot = vstack(hstack(paccTr,paccTe), hstack(ppcTr,ppcTe), hstack(prmseTr,prmseTe))
```
